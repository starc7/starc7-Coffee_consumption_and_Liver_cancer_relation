**Coffee and Live cancer corealtion**

This project is to find the relation between coffee consumption and the occurences of  liver cancer in different countries.

**Data sources:**

Data till 2020

Coffee consumption data : https://worldpopulationreview.com/country-rankings/coffee-consumption-by-country

Liver cancer data:
https://www.wcrf.org/cancer-trends/liver-cancer-statistics



```python
import pandas as pd
```

```python
coffee_data = pd.read_csv('coffee_consumption.csv')
liver_data = pd.read_csv('liver.csv')
```

```python
merged_data = pd.merge(coffee_data, liver_data, on='country')
```

```python
df = merged_data[['country', 'coffeeConsumptionByCountry_perCapitaCons2020','ASR/1000']]
df.rename(columns = {'coffeeConsumptionByCountry_perCapitaCons2020':'coffeeConsumption'}, inplace = True)
print(df)
```

```python
correlation_coefficient = df['coffeeConsumption'].corr(df['ASR/1000'])
print("Correlation Coefficient:", correlation_coefficient)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['coffeeConsumption'], merged_data['ASR/1000'], color='blue', alpha=0.7)
plt.title('Coffee Consumption vs. Liver Cancer Rates')
plt.xlabel('Coffee Consumption (cups per capita per year)')
plt.ylabel('Liver Cancer Rates (per 1000 people)')
plt.grid(True)
plt.show()
```

**Insights**

The correlation_coefficient is 0.3298065935460454. (Round off: 0.33)

The correlation coefficient indicates a moderate positive correlation between coffee consumption and liver cancer rates. This suggests that, on average, as coffee consumption increases in a country, liver cancer rates tend to increase as well. However, the correlation coefficient of 0.33 suggests that the relationship between coffee consumption and liver cancer rates is not extremely strong. While there is a discernible trend, other factors likely contribute significantly to liver cancer rates.
